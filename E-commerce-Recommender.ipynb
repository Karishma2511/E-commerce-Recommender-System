```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

data = pd.read_csv('test.csv', encoding='latin-1')

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
data['product_title'].value_counts().head(10).plot(kind='barh')
plt.title('Top 10 Most Common Product Titles')
plt.xlabel('Count')
plt.ylabel('Product Title')
```

```python
plt.subplot(1, 2, 2)
data['search_term'].value_counts().head(10).plot(kind='barh', color='orange')
plt.title('Top 10 Most Common Search Terms')
plt.xlabel('Count')
plt.ylabel('Search Term')
plt.tight_layout()
plt.show()
```

```python
product_title_cloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(data['product_title']))
search_term_cloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(data['search_term']))
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.imshow(product_title_cloud, interpolation='bilinear')
plt.axis('off')
plt.title('Product Titles')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(search_term_cloud, interpolation='bilinear')
plt.axis('off')
plt.title('Search Terms')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('test.csv', encoding='latin-1')
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(y='product_title', data=data, order=data['product_title'].value_counts().iloc[:10].index)
plt.title('Count of Top 10 Product Titles')
plt.xlabel('Count')
plt.ylabel('Product Title')
```

```python
plt.subplot(1, 2, 2)
sns.countplot(y='search_term', data=data, order=data['search_term'].value_counts().iloc[:10].index)
plt.title('Count of Top 10 Search Terms')
plt.xlabel('Count')
plt.ylabel('Search Term')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
unique_products = data['product_uid'].nunique()
unique_search_terms = data['search_term'].nunique()
plt.bar(['Unique Products', 'Unique Search Terms'], [unique_products, unique_search_terms], color=['blue', 'orange'])
plt.title('Number of Unique Products and Search Terms')
plt.ylabel('Count')
plt.show()
```
